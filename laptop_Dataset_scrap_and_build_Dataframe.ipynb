```python
with open(path,'r', encoding='utf-8') as f:
    html = f.read()
```

```python
# Import library

from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
# load soup
soup = BeautifulSoup(html , 'lxml')
```

```python
# load chunk for check what to fetch

container =soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
container[6]
```

```python
# name,price,ratings, processor, core, ram, hdd, gpu, resolution, os, warranty, score

# price = class="price">₹16,990</span>

# ratings = <span C style="--rating: 4.75;">

# names =    <a class="name clamp-2" data-target="_blank" data-way="" 
# href="/laptops/primebook-4g-android-laptop-mediatek-kompanio-ppd1n0eewaga">
# <h2>Primebook 4G Android Laptop (MediaTek Kompanio 500/ 4GB/ 64GB eMMC/ Android 11)</h2></a>



#  specs = <ul class="sm-feat specs">
#         <!--[-->
#         <li class="">MediaTek MTK8788</li> processor 
#         <li class="">Octa Core</li>  Core
#         <li class="">4 GB LPDDR4 RAM</li> ram
#         <li class="">64 GB Hard Disk</li> hdd
#         <li class="">ARM Mali G72</li> GPU
#         <li class="">11.6 inches, 1366 x 768 pixels</li> resolution
#         <li class="">Android 11 OS</li> os
#         <li class="">1 Year Warranty</li> warranty
        
        
# <div class="score rank-4-bg"><b>26</b><small>Spec Score</small></div> score
```

```python
# Scrap data from html file 

laptop_name = []
laptop_price = [] 
laptop_ratings = []
laptop_ratings1 = []
laptop_ratings2 = []
laptop_processor = [] 
laptop_processor_core = [] 
laptop_ram = [] 
laptop_memory = [] 
laptop_gpu = [] 
laptop_resolution = [] 
laptop_os = []
laptop_warranty = []


for i in soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}):
    try:
        laptop_name.append(i.find('h2').text)
    except:
        laptop_name.append(np.nan)
        
        
    try:
        laptop_price.append(i.find('span',{'class':'price'}).text)
    except:
        laptop_price.append(np.nan)    
        
    try:
        laptop_ratings2.append(i.find('div',{'class':'score rank-4-bg'}).find('b').text)
    except:
        # laptop_ratings2.append(np.nan) 
        # np.nan create issue of array index not match
        laptop_ratings2.append("0") 
        
        
    try:
        laptop_ratings.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        # laptop_ratings.append(np.nan) 
        # np.nan create issue of array index not match
        laptop_ratings.append("0") 
        
        
    try:
        laptop_ratings1.append(i.find('div',{'class':'score rank-3-bg'}).find('b').text)
    except:
        # laptop_ratings1.append(np.nan)    
        # np.nan create issue of array index not match
        laptop_ratings1.append("0")    
        
        
        
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        laptop_processor.append(x[0].text)
    except:
        laptop_processor.append(np.nan)
        
        
    try:
        laptop_processor_core.append(x[1].text)
    except:
        laptop_processor_core.append(np.nan)
        
        
    try:    
        laptop_ram.append(x[2].text)
    except:
        laptop_ram.append(np.nan)
        
        
    try:
        laptop_memory.append(x[3].text)
    except:
        laptop_memory.append(np.nan)
        
        
    try:
        laptop_gpu.append(x[4].text)
    except:
        laptop_gpu(np.nan)
        
        
    try:
        laptop_resolution.append(x[5].text)
    except:
        laptop_resolution.append(np.nan)
        
        
    try:
        laptop_os.append(x[6].text)
    except:
        laptop_os.append(np.nan)
        
        
    try:
        laptop_warranty.append(x[7].text)
    except:
        laptop_warranty.append(np.nan)
        
            
```

```python
# length check

print(len(laptop_name))
print(len(laptop_price))
print(len(laptop_ratings))
print(len(laptop_ratings1))
print(len(laptop_ratings2))
print(len(laptop_processor)) 
print(len(laptop_processor_core))
print(len(laptop_ram)) 
print(len(laptop_memory))
print(len(laptop_gpu)) 
print(len(laptop_resolution))
print(len(laptop_os))
print(len(laptop_warranty))
```

```python
# DataFrame Created

df = pd.DataFrame({
    'model':laptop_name,
    'price':laptop_price,
    'rating':laptop_ratings,
    'rat1':laptop_ratings1,
    'rat2':laptop_ratings2,
    'processor':laptop_processor, 
    'processor_core':laptop_processor_core,
    'ram':laptop_ram,
    'memory':laptop_memory, 
    'gpu':laptop_gpu,
    'resolution':laptop_resolution, 
    'os':laptop_os,
    'warranty':laptop_warranty
})
```
